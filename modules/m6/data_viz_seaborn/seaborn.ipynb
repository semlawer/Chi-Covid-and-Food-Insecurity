### Introduction to Seaborn 
What is Seaborn? 
 - A library for making staistical graphics in Python. 
 - It's built on top of matplotlib (Python's main plotting library). 
 - Why use Seaborn over the bultin-in matplotlib library? 
     - Provide more concise control over your plots styling with several built-in themes
     - Works closely with pandas data structures unlike matplotlib. 
     
Seaborn provides many different types of plots to help visualize your data. However, we will focus primilary on generating plots for: 
    - Examining relationships between multiple variables in a 
      dataset. 
    - Visulazing data that uses categorical variables to show 
      observations or aggregate statistics.

### Seaborn Basics 

First, we'll take a look at using ```scatterplot``` to create a scatter plot of some data. A scatter plot uses points to show the relationship between two sets of data. We simply need to define the values for ```x``` and ```y```variables for the graph. 

````python
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
sns.scatterplot(x=height, y=weight)
plt.show()
````

Each plotting funciton in Seaborn returns a matplotlib axisgrid object. This object includes additional methods for adding content to a plot: 
   - ```set_title(str)```   : inserts a title for the plot 
   - ```set_ylabel(str)``` : set the labels of the y-axis 
   - ```set_xlabel(str)``` : set the labels of the x-axis 
   - ```set_xlim(lower,upper)``` : set the x-axis view limits 
   - ```set_ylim(lower,upper)``` : set the y-axis view limits 

````python
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
axes = sns.scatterplot(x=height, y=weight)
axes.set_title("Height/Weight Plot")
axes.set_ylabel("Weight")
axes.set_xlabel("Height")
plt.show()
````

**What if you want multiple plots on one figure?**

Use the ``f, axes = plt.subplots(nRows, nCols)``. The ``subplots`` function will return a figure (``f``) that has the ability to contain ``nRows*nCols`` plots. The ``axes`` is matplotlib axisgrid object list that contains references to the plots to be added to figure. Use the ``ax`` keyword argument from the plotting functions to assign axes object from the list. This will allow the plots to be shown in the figure. 

````python
# Data for the plots 
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
animal_counts = [1,4,5,6]
animals = ["Cat", "Dog", "Fish", "Bird"]
ages = [34, 23, 54, 75]
names = ["Bob", "Sally", "Joe", "Jane"]


# Create the axes and figure 
f, axes = plt.subplots(1, 3) # Has space for two plots 1*3 

# Create the scatterplots. Assign the ax keyword argument to a
# Axis object from the axes lists 
sns.scatterplot(x=height, y=weight, ax=axes[0])
sns.scatterplot(x=animals, y=animal_counts, ax= axes[1])

axes[0].set_title("Weight Height Plot")
axes[0].set_ylabel("Weight")
axes[0].set_xlabel("Height")

axes[1].set_title("Animal Plot")
axes[1].set_ylabel("Counts")

axes[1].set_xlabel("Animal")

sns.scatterplot(x=names, y=ages, ax = axes[2])
axes[2].set_title("Names Ages Plot")
axes[2].set_ylabel("Ages")
axes[2].set_xlabel("Names")


f.suptitle("My ScatterPlots")


plt.show()
````

Finally, you can save the figure to a file using the ``savefig`` function. 

````python
# Data for the plots 
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
animal_counts = [1,4,5,6]
animals = ["Cat", "Dog", "Fish", "Bird"]

# Create the axes and figure 
f, axes = plt.subplots(1, 2) # Has space for two plots 1*2 

# Create the scatterplots. Assign the ax keyword argument to a
# Axis object from the axes lists 
sns.scatterplot(x=height, y=weight, ax=axes[0])
sns.scatterplot(x=animals, y=animal_counts, ax= axes[1])

axes[0].set_title("Weight Height Plot")
axes[0].set_ylabel("Weight")
axes[0].set_xlabel("Height")

axes[1].set_title("Animal Plot")
axes[1].set_ylabel("Counts")
axes[1].set_xlabel("Animal")

f.suptitle("My ScatterPlots")

plt.savefig("scatterplots.png")
plt.show()
````

### Creating Multiple Figures 
Use the following functions to clear the matplotlib window to show/create multiple figures: 

- ``plt.cla()``   : clear the current axis 
- ``plt.clf()``   : clear an entire figure 
- ``plt.close()`` : close the a window 

````python
def figure1():
    height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
    weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
    sns.set(style="whitegrid")
    axes = sns.scatterplot(x=height, y=weight)
    axes.set_title("Height/Weight Plot")
    axes.set_ylabel("Weight")
    axes.set_xlabel("Height")
    plt.savefig("scatterplots-fig1.png")

def figure2():
    animal_counts = [1,4,5,6]
    animals = ["Cat", "Dog", "Fish", "Bird"]
    sns.set(style="darkgrid")
    axes = sns.scatterplot(x=animals, y=animal_counts)
    axes.set_title("Animal Plot")
    axes.set_ylabel("Counts")
    axes.set_xlabel("Animal")
    plt.savefig("scatterplots-fig2.png")

figure1() 
plt.show()
plt.clf()
figure2()
plt.show()
````

Another way to create a ```scatterplot``` is to use the ```relplot``` function. For Seaborn, ```relplot()``` allows you to define different type of plots using a single function. By defining the ```kind``` keyword agrument, you can then specify the type of plot, which could be line, bar, or etc. However, the data must be specified within a DataFrame and the ```x``` and ```y``` variables need to specify a Series within the DataFrame. 

````python
import pandas as pd 
import numpy as np 
col_data = {"height":height, "weight":weight}
pd_data = pd.DataFrame(data=col_data)
sns.relplot(x="height", y="weight", data=pd_data)
````

### Visualizing statistical relationships
From Seaborn,  
  " Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship." 
  
- To see how variables within data are related with each other, we will be using
    - ```scatterplots```- allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between the variables within a dataset. Not always most optimal. 
    
    - ```lineplots``` -  good for datasets where you need to understand changes in one variable as a function of time, or a continuous variable.
    

````python
tips = sns.load_dataset("tips") # tips is a Pandas dataframe 
tips
````

````python
sns.relplot(x="total_bill", y="tip", data=tips)
````

````python
# Another dimension can be added to the plot by coloring 
# the points according to a third variable. In seaborn, 
# this is referred to as using a “hue semantic”, because the 
# color of the point gains meaning. 
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips) 
````

````python
# To emphasize the difference between the classes, 
# and to improve accessibility, you can use a different 
# marker style for each class.
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips)
````

````python
# Can add four variables to the chart using both 
# hue & style (but it can get messy)
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips)
````

````python
# The third kind of semantic variable changes the size of each point:
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
````

### Color Themes 

Seaborn allows for you to change the color theme for the different levels of the hue variable. Use the ``palette`` keyword argument to specify the color theme to be used for a specific type of plot. You need to specify a string that represents a color theme. Examples include: ``"deep"``, ``"muted"``, ``"pastel"``, ``"bright"``, ``"BuGn_r"`` (gradient blues and greens), and ``"Blues"`` (gradient blues).

**Note**: You can also define your own color themes and modify the current default color theme. Look in the Seaborn documentation for further details. 

````python
sns.relplot(x="total_bill", y="tip", hue="smoker", palette="BuGn_r",  data=tips)
````

### Lineplots 

````python
# can use lineplot() or replot with kind="line"
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
````

````python
fmri = sns.load_dataset("fmri")
fmri
````

````python
# Complex datasets will have multiple measurements for the same value of the x variable. 
# The default behavior in seaborn is to aggregate the multiple measurements at each x value by plotting 
# the mean and the 95% confidence interval around the mean
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
````

````python
# The confidence intervals are computed using bootstrapping, 
# which can be time-intensive for larger datasets. It’s therefore possible to disable them
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri)
````

````python
# To turn off aggregation altogether, set the estimator parameter to None 
# This might produce a strange effect when the data have multiple observations at each point.
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri)
````

````python
# Use the hue and style keywords to plot more than one variable
sns.relplot(x="timepoint", y="signal", hue="event",  kind="line", 
            data=fmri)
````

### Plotting with categorical data

In Seaborn, there are many ways to visualize a relationship involving categorical data. We will look at two common ways: 

- ``stripplot()`` : categorical scatter plots 
- ``barplot()``   : operates on a full dataset and applies a function to obtain the estimate (taking the mean by default).

- ``countplot()`` : shows the number of observations in each category

Similiar to the relational plotting function ``relplot``. There is a categorical function (``catplot``) that allows for generating different types of categorical plots. Again you specify the ``kind`` keyword argument to designate the type of categorical plot to generate.  

````python
tips = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tips);
````

````python
# When there are multiple observations in each category, 
# it also uses bootstrapping to compute a confidence interval around the estimate, 
# which is plotted using error bars
titanic = sns.load_dataset("titanic")
sns.barplot(x="sex", y="survived", data=titanic)
````

````python
sns.countplot(x="deck", palette="ch:.25", data=titanic);
````
